```python
data=pd.read_excel("Boston Housing Dataset 1978.xlsx")
data
```

```python
X=data.iloc[:,:-1].values
y=data['MEDV'].values
```

```python
import numpy as np
ones=np.ones([506,1])
X = np.hstack((ones, X))
```

```python
import statsmodels.api as sm
```

```python
#MODELO DE REGRESION MULTIPLE

ols=sm.OLS(y,X)
ols_results=ols.fit()
ols_results.summary()
```
